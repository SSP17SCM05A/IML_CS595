```python
def prob(df,target,w):
    pred=0
    for i in range(df.shape[0]):
        s=np.dot(df[i],w).sum()
        if target[i]:
            pred+=1-(1./(1 + np.exp(s)))
        else:
            pred+=1./(1 + np.exp(s))
            return pred
    
```
